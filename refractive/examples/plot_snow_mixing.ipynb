```python
Temperature = np.array(260.0)
#Frequencies = np.linspace(1,250,500)
Frequency = np.array(94)
densities = np.linspace(0,snow.ice_density,1000)

snow_b = snow.n(Temperature,Frequency,densities,model_mix='bruggeman')
snow_mga = snow.n(Temperature,Frequency,densities,model_mix='maxwell_garnett')
snow_mgi = snow.n(Temperature,Frequency,densities,model_mix='maxwell_garnett')
```

```python
f, ax = plt.subplots()
s = ax.scatter(snow_b.imag, snow_b.real,marker='.', c=densities,label='bruggeman')
s = ax.scatter(snow_mga.imag, snow_mga.real,marker='.', c=densities,label='maxwell-garnett air inclusions')
#s = ax.scatter(snow_mgi.imag, snow_mgi.real, c=densities,label='maxwell-garnett ice inclusions')
ax.legend()
ax.grid()
cbar = f.colorbar(mappable=s, ax=ax)
cbar.set_label('Snow density    [g/cm$^3$]')
ax.set_ylabel('Real part')
ax.set_xlabel('Imaginary part')
ax.set_title('Refractive index')
ax.set_xlim([0,0.002])
```
